```python
import pywindow as pw
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import os
```

```python
cages = np.loadtxt("all_cages/all_cages.txt", dtype=str)
```

```python
property_file = open("pywindow_cage_descriptors.csv", "w")
property_file.write("cage,molecule_diameter_A,pore_diameter_A,nb_windows,window1,window2,window3,window4,window5,window6,window7,window8,window9,window10,\n")
```

```python
for cage in cages:
    print("Cage ", cage)
    molsys = pw.MolecularSystem.load_file("all_cages/%s.xyz" % cage)
    mol = molsys.system_to_molecule()
    mol.calculate_maximum_diameter()
    mol.calculate_pore_diameter()
    mol.calculate_windows()
    nb_windows = len(mol.calculate_windows())
    property_file.write("%s,%f,%f,%d" % (cage,
                       mol.properties['maximum_diameter']['diameter'],
                       mol.properties['pore_diameter']['diameter'],
                       nb_windows)
                       )
    for i in range(nb_windows):
        if i >= nb_windows:
            property_file.write(",%f" % (-1))
        else:
            property_file.write(",%f" % (mol.properties['windows']['diameters'][i]))
    property_file.write("\n")
property_file.close()
```

```python
inaccessible_cages = np.genfromtxt('inaccessible_cages.txt',dtype='str')
df = pd.read_csv("pywindow_cage_descriptors.csv")
print("Cage, Maximum Window Diameter")
for cage in inaccessible_cages:    
    diameters = np.array(df[df["cage"] == cage].values[0][4:], dtype='float64')
    diameters = diameters[~np.isnan(diameters)]
    maxi = np.max(diameters)
    print("{}: {}".format(cage, maxi))
```
