```python
import pywindow as pw
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import os
```

```python
cages = np.loadtxt("all_cages/all_cages.txt", dtype=str)
```

```python
flexible_cages = np.loadtxt("all_cages/flexible_snapshots/cc3_files.txt", dtype=str)
```

```python
property_file = open("pywindow_cage_descriptors.csv", "w")
property_file.write("cage,molecule_diameter_A,pore_diameter_A,nb_windows\n")
```

```python
for cage in cages:
    print("Cage ", cage)
    molsys = pw.MolecularSystem.load_file("all_cages/%s_aligned.xyz" % cage)
    mol = molsys.system_to_molecule()
    mol.calculate_maximum_diameter()
    mol.calculate_pore_diameter()
    nb_windows = len(mol.calculate_windows())
    property_file.write("%s,%f,%f,%d\n" % (cage,
                       mol.properties['maximum_diameter']['diameter'],
                       mol.properties['pore_diameter']['diameter'],
                       nb_windows)
                       )
property_file.close()
```

```python
flex_property_file = open("flex_pywindow_cage_descriptors.csv", "w")
flex_property_file.write("cage,molecule_diameter_A,pore_diameter_A,nb_windows,window1,window2,window3,window4\n")
```

```python
for cage in flexible_cages:
    print("Cage ", cage)
    molsys = pw.MolecularSystem.load_file("all_cages/%s" % cage)
    mol = molsys.system_to_molecule()
    mol.calculate_maximum_diameter()
    mol.calculate_pore_diameter()
    mol.calculate_windows()
    nb_windows = len(mol.calculate_windows())
    flex_property_file.write("%s,%f,%f,%d" % (cage,
                       mol.properties['maximum_diameter']['diameter'],
                       mol.properties['pore_diameter']['diameter'],
                       nb_windows)
                       )
    for i in range(nb_windows):
        flex_property_file.write(",%f" % (mol.properties['windows']['diameters'][i]))
    flex_property_file.write("\n")
    
flex_property_file.close()
```

```python
df = pd.read_csv("flex_pywindow_cage_descriptors.csv")
df
```

```python
fig = plt.figure(figsize=(10,8))
windows = np.concatenate([df['window1'].values, df['window2'].values, df['window3'].values, df['window4'].values])
# One image didn't work...
myrange = [i for i in range(len(windows)) if i != 2378]
windows = windows[myrange]
len(windows)
plt.hist(windows, bins=75)
plt.xlabel('Pore diameter (Å)')
plt.ylabel('Number of configurations')
plt.xlim(2.2, 5.2)
plt.savefig("CC3_histogram.png", dpi=300, format="png")
```
